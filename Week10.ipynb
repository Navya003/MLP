 Load the wine dataset from sklearn.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

df=load_wine()

X_train, X_test, y_train, y_test=train_test_split(df.data, df.target, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
adc=AdaBoostClassifier(random_state=1)
adc.fit(X_train, y_train)
print('Score on Training set:', adc.score(X_train, y_train))
print('Score on Test set:', adc.score(X_test, y_test))
```
