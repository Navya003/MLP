Q2: During which month did the highest number of flights occur? Months are represented by numerical codes, with January corresponding to 1, February to 2, and so forth.

```python
# Grouping by 'Month' and counting the number of flights in each month
flights_by_month = df.groupby('Month').size()

# Finding the month with the highest number of flights
max_flights_month = flights_by_month.idxmax()

# Print the result
print("Month with the Highest Number of Flights:", max_flights_month)
```

Q3: Is the average price of flight tickets higher on weekends (Saturday and Sunday) or on weekdays (Remaining 5 days)?

```python
# 'WeekDay' column to identify weekdays and weekends
#  `isin` function to filter based on weekdays and weekends
weekday_prices = df.loc[~df['WeekDay'].isin(['Saturday', 'Sunday']), 'Price']
weekend_prices = df.loc[df['WeekDay'].isin(['Saturday', 'Sunday']), 'Price']

# Calculate the average price for weekdays and weekends
average_price_weekdays = weekday_prices.mean()
average_price_weekends = weekend_prices.mean()

# Print the results
print("Average Price on Weekdays:", average_price_weekdays)
print("Average Price on Weekends:", average_price_weekends)
```

Q4:
Two of the entries in the 'Additional_Info' column are 'No info' and 'No Info'. Replace all occurrences of 'No Info' with 'No info'. How many flights fall under airline 'IndiGo' and have 'No info' as additional information?

```python
# Replace 'No Info' with 'No info' in the 'Additional_Info' column
df['Additional_Info'] = df['Additional_Info'].replace('No Info', 'No info')

# Count the number of flights that are 'IndiGo' and have 'No info' as additional information
indiGo_no_info_flights = df[(df['Airline'] == 'IndiGo') & (df['Additional_Info'] == 'No info')]

# Get the count of such flights
num_indiGo_no_info_flights = len(indiGo_no_info_flights)

# Print the result
print("Number of IndiGo flights with 'No info' as additional information:", num_indiGo_no_info_flights)
```

Q5:
Convert the values of 'Duration' into seconds. Enter the average duration (in seconds) of a flight. Enter your answer correct to two decimal places.

```python
# Convert 'Duration' to seconds
df['Duration'] = pd.to_timedelta(df['Duration']).dt.total_seconds()

# Calculate the average duration in seconds
average_duration_seconds = df['Duration'].mean()

# Print the result
print("Average Duration of a flight (in seconds): {:.2f}".format(average_duration_seconds))
```

Apply the following functions to the columns Dep_Time and Arrival_Time:

Transform the values in the 'dep_time' and 'arrival_time' columns to represent the hour component. For instance, if an entry is 10:05 June 13 or 10:05, the corresponding value should be 10.
Then convert the time into four categories as follows:
5 <= hour < 12 = Morning
12 <= hour < 17 = Afternoon
17 <= hour < 20 = Evening
20 <= hour < 5 = Night

Note: Please ensure that you make the changes directly within the dataset and continue to use that modified dataset for subsequent questions.

```python
# Function to extract the hour component and categorize into Morning, Afternoon, Evening, Night
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 20:
        return 'Evening'
    else:
        return 'Night'

# Transform 'Dep_Time' and 'Arrival_Time' to represent the hour component
df['Dep_Hour'] = pd.to_datetime(df['Dep_Time']).dt.hour
df['Arrival_Hour'] = pd.to_datetime(df['Arrival_Time']).dt.hour

# Apply the categorize_time function to create categories for departure time
df['Dep_Time_Category'] = df['Dep_Hour'].apply(categorize_time)

# Apply the categorize_time function to create categories for arrival time
df['Arrival_Time_Category'] = df['Arrival_Hour'].apply(categorize_time)

df.head()
```

Q6:
How many flights started in the Morning and arrived the destination at Evening?
